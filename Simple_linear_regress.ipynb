```python

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
# Sample dataset (Logarithmic growth)
# X
study_hours = np.array([1, 2, 3, 4, 5, 6, 7]).reshape(-1, 1)
# Y
exam_scores = np.array([50, 65, 75, 82, 87, 90, 92])
# Transforming X using log for Logarithmic regression
study_hours_log = np.log(study_hours)
# Fitting the model
model = LinearRegression()
model.fit(study_hours_log.reshape(-1, 1), exam_scores)
# Predictions
predicted_scores = model.predict(study_hours_log.reshape(-1, 1))
# Plotting
plt.scatter(study_hours, exam_scores, color='blue', label='Actual Data')
plt.plot(study_hours, predicted_scores, color='red', label='Logarithmic Regression Curve')
plt.xlabel('Study Hours')
plt.ylabel('Exam Score')
plt.legend()
plt.show()
# Performance metrics
print("Mean Squared Error:", mean_squared_error(exam_scores, predicted_scores))
print("R-squared:", r2_score(exam_scores, predicted_scores))
```
