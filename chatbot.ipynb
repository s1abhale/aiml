```python
import re

def chatbot():

    print("Chatbot: Hello! How can I assist you? Type 'bye' to exit.")

    while True:

        user_input = input("You: ")

        if user_input.lower() == 'bye':

            print("Chatbot: Goodbye! Have a great day!")

            break

        elif re.search(r'hi|hello|hey', user_input, re.IGNORECASE):

            print("Chatbot: Hello! How can I help you?")

        elif re.search(r'weather', user_input, re.IGNORECASE):

            print("Chatbot: The weather is great today!")

        elif re.search(r'name', user_input, re.IGNORECASE):

            print("Chatbot: I'm a simple chatbot. What's your name?")

        else:

            print("Chatbot: I didn't understand that.")

chatbot()
```
