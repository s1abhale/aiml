```python
import numpy as np
import statistics as stats
# Sample data
data = [12, 15, 12, 15, 19, 21, 56]
# Central Tendency
mean = np.mean(data)
median = np.median(data)
mode = stats.mode(data)
# Dispersion
variance = np.var(data)
std_dev = np.std(data)
# Output
print("Mean:", mean)
print("Median:", median)
print("Mode:", mode)
print("Variance:", variance)
print("Standard Deviation:", std_dev)
```
